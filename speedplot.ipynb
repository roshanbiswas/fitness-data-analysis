```python
data_needed = data.drop(['Average weight (kg)','Max weight (kg)','Min weight (kg)','Biking duration (ms)','Calisthenics duration (ms)','Hiking duration (ms)','Sleep duration (ms)','Inactive duration (ms)'], axis=1)
```

```python
print(data_needed.tail())
```

```python
print(data_needed.plot(y = 'Average speed (m/s)'))
```

```python
print(data_needed["Average speed (m/s)"])
```

```python
new_list = []
for n in data_needed["Average speed (m/s)"]:
    if (n<2.5):
        new_list.append(n)
new_list
```

```python
len( new_list)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
plt.plot(new_list,color='red',linestyle='dashed', linewidth=3) # for color : plt.plot(new_list,color='red')
plt.title('Average speed of Roshni(m/s)', fontsize= 16) # plt.title('your title')
plt.xlabel('speed') # plt.xlabel('x axis labelling')
plt.ylabel('time') # plt.ylabel('y axis labelling')
plt.savefig('avg-speed-plot.png')
```

1. how fast can roshni walk?

```python
roshni_max_speed = max(new_list)
print("Roshni's max speed is:", ('%.3f' % roshni_max_speed), "m/s.") # for rounding off decimal points use - round(n, no. of decimal digits you want), -#
```

2. how slow can roshni walk?

```python
roshni_min_speed = min(new_list)
print("Roshni's min speed is:", ('%.3f' % roshni_min_speed), "m/s.")
```

3. how much distance does roshni walk in a month on average?

```python
roshni_avg_speed = sum(new_list)/len(new_list)
roshni_monthly_dist = (roshni_avg_speed)*(30*24*3600)
roshni_monthly_dist_km = roshni_monthly_dist/1000
```

```python
print("Roshni walks a distance of", ('%.3f' % roshni_monthly_dist_km), "km in a month on average.")
```
